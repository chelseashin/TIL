```python
# 네이버 모바일 증권 스크래핑

import requests
from bs4 import BeautifulSoup

url = "https://m.stock.naver.com/marketindex/index.nhn"
req = requests.get(url).text
soup = BeautifulSoup(req, 'html.parser')
rate = soup.select('#marketindexLastList > li')

for i in rate:
    if i.select_one('.stock_item') != None:
        item = i.select_one('.stock_item').text
        price = i.select_one('.stock_price').text
        print(item, price)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://m.stock.naver.com/marketindex/index.nhn"
req = requests.get(url).text
soup = BeautifulSoup(req, 'html.parser')
items = soup.select('#marketindexLastList > li')

for li in items:
    if li.select_one('.stock_item') != None:
        print(li.select_one('.stock_item').text)
        print(li.select_one('.stock_price').text)
```

```python
# 다음 모바일 환율 스크래핑

daum = "http://m.exchange.daum.net/mobile/exchange/exchangeMain.daum"
req = requests.get(daum).text
soup = BeautifulSoup(req, 'html.parser')

for point in soup.select('.link'):
    name = point.select_one('.name').text
    money = point.select_one('.idx').text
    print(name, money)
```
