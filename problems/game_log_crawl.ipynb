```python
games_list=[]
for year in range(2010,2019):
    for month in [3,4,5,6,7,8,9,10]:
        driver.get('https://sports.news.naver.com/kbaseball/schedule/index.nhn?&month=0'+str(month)+'&year='+str(year))
        for i in driver.find_elements_by_css_selector("span.td_btn"):
            games_list.append(i.find_elements_by_css_selector("a")[0].get_attribute('href'))
    print(year)
```

```python
for i in range(0,len(games_list))
driver.get(games_list[i])
```

```python
bat_res=[]
game_date=[]
game_time=[]
for j in range(len(games_list)):
    driver.get("https://sports.news.naver.com/gameCenter/textRelay.nhn?gameId"+games_list[j].split("gameId")[1])
    time.sleep(5)


    try:
        driver.find_element_by_id("inning_tab_all").click()
    except:
        print("에러 ㅅㅂ")
        continue

    bat_res1=[]

    for i in driver.find_element_by_id("relay_text").find_elements_by_css_selector("p"):
        bat_res1.append(i.text)
    bat_res1=list(reversed(bat_res1))

    bat_res.append(bat_res1)

    game_date.append(driver.find_element_by_css_selector("strong.t_num").text)
    game_time.append(driver.find_element_by_css_selector("span.t_num.time").text)
    print(j)
```
