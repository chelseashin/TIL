## DBSCAN Clustering
#### Density-Based Clustering Algorithm

* 타 알고리즘에 비해 빠름
* noise 다루기 쉬움
* 두 개의 parameter : Epsilon, MinPls
* Directly density-reachable
* density-connected

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.05, min_samples=20)
y_pred = dbscan.fit_predict(X)
print(y_pred[:10])
```

```python
for i, (eps, min_samples) in enumerate([(0.05, 20), (0.05, 20), (0.06, 15), (0.06, 6)]):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    y_pred = dbscan.fit_predict(X)
    plt.figure(i+1, figsize=(5, 5))
    plt.scatter(X[:, 0], X[:, 1],
               cmap="tab10", s = 4, c=y_pred)
    plt.title("eps: {}, min_samples: {}".format(eps, min_samples))
plt.show()
```
